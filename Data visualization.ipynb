```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

# 1. Microsoft Stock data

```python
df_stock = pd.read_csv('data/Microsoft_Stock.csv')
df_stock.info()
```

```python
dates_stock = np.array(pd.to_datetime(df_stock['Date']))
volumes = np.array(df_stock['Volume']).astype('float32')
close_prices = np.array(df_stock['Close']).astype('float32')
```

## Volume per day

```python
plt.figure(figsize=(25, 10), dpi=80)
plt.plot(dates_stock,volumes, linewidth=3)
plt.xlabel('Time', fontweight='bold', fontsize=22)
plt.ylabel('Stock Volume', fontweight='bold', fontsize = 22)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('images/MS_Stock_volume.png')
plt.show()
```

## Close price

```python
plt.figure(figsize=(25, 10), dpi=80)
plt.plot(dates_stock,close_prices, linewidth=3)
plt.xlabel('Time', fontweight='bold', fontsize=22)
plt.ylabel('Close price (USD)', fontweight='bold', fontsize = 22)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('images/MS_Stock_close_price.png')
plt.show()
```

# 2. Electricity Consumption

```python
df_elec = pd.read_csv('data/Electric_Production.csv')
df_elec.info()
```

```python
dates_elec = np.array(pd.to_datetime(df_elec['DATE']))
production = np.array(df_elec['value']).astype('float32')
```

```python
plt.figure(figsize=(25, 10), dpi=80)
plt.plot(dates_elec,production, linewidth=3)
plt.xlabel('Time', fontweight='bold', fontsize=22)
plt.ylabel('Electricity Production', fontweight='bold', fontsize = 22)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig('images/Electric_Production.png')
plt.show()
```
