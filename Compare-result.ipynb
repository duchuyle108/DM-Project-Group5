```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Electricity Production data

```python
elec_real = np.loadtxt("output/Elec_real.csv")
elec_lstm = np.loadtxt("output/Elec_lstm.csv")
elec_gru = np.loadtxt("output/Elec_gru.csv")
elec_arima = np.loadtxt("output/Elec_arima.csv")
```

```python
plt.figure(figsize=(18, 10), dpi=80)
plt.plot(elec_real, label = "Real values", marker='o', linewidth=2)
plt.plot(elec_lstm, label = "LSTM", marker='s', color = 'g', linewidth=4)
plt.plot(elec_lstm, label = "GRU", marker='v', color = 'black', linewidth=2)
plt.plot(elec_arima, label = "ARIMA", marker='^', color = 'orange', linewidth=2)
plt.xlabel('Timesteps', fontweight='bold', fontsize=20)
plt.ylabel('Electricity Production', fontweight='bold', fontsize = 20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=18)
plt.savefig("images/compare_elec.png")
plt.show()
```
